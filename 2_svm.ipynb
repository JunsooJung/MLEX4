```python
import numpy as np
import matplotlib.pyplot as plt

points = np.array([[1, 9], [9, 5], [5, 2]])
labels = np.array([1, 1, -1])
positive_points = points[labels == 1]
negative_points = points[labels == -1]

plt.figure(figsize=(8, 8))
plt.scatter(positive_points[:, 0], positive_points[:, 1], color='blue', label='Label (1)', s=100, edgecolor='k')
plt.scatter(negative_points[:, 0], negative_points[:, 1], color='red', label='Label (-1)', s=100, edgecolor='k')
for i, point in enumerate(points):
    x, y = point
    plt.annotate(f"({x}, {y})", (x, y), textcoords="offset points", xytext=(5, 10), ha='center')
plt.axhline(0, color='black',linewidth=1, linestyle='solid')
plt.axvline(0, color='black',linewidth=1, linestyle='-')
plt.grid(color = 'gray', linestyle = '--', linewidth = 1)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.xlim(-1, 10)
plt.ylim(-1, 10)
plt.show()
```
